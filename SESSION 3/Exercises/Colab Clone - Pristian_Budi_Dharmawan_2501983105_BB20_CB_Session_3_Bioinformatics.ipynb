```python
# Install BioPy
%pip install biopython
# Reading PDB Files
from Bio.PDB.PDBParser import PDBParser
import os
```

```python
# Create a parser
parser = PDBParser()
pdb_file1 = os.path.abspath("/content/mmdb_6LU7.pdb")
pdb_file2 = os.path.abspath("/content/6lu7.pdb")
structure = parser.get_structure(pdb_file1, pdb_file2)
structure
```

```python
model  = structure[0]
```

```python
# Check for chains
for chain in model:
    print(f'Chain {chain},Chain_ID{chain.id}')
```

```python
# Check for Residue
for chain in model:
    print(f'Chain {chain},Chain_ID{chain.id}')
    for residue in chain:
        print(residue)
```

```python
# Check for atoms
for chain in model:
    print(f'Chain {chain},Chain_ID{chain.id}')
    for residue in chain:
        for atom in residue:
            print(atom)
```
